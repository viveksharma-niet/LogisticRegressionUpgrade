```python
import numpy as np
import pandas as pd

data = pd.read_csv('https://raw.githubusercontent.com/viveksharma-niet/LogisticRegressionUpgrade/refs/heads/main/AR%20-%20Examples%20-%201.6.csv')
data.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data.info()
```

```python
sns.scatterplot(data, y='Marks', x='number_courses')
plt.show()
```

```python
sns.scatterplot(data, y='Marks', x='time_study')
plt.show()
```

```python
y = data['Marks']
X = data.drop('Marks', axis=1)
```

```python
X.head()
```

```python
from sklearn.linear_model import LinearRegression

lm = LinearRegression()
lm.fit(X, y)
```

```python
y_pred = lm.predict(X)
```

```python
error = y-y_pred
```

```python
sns.scatterplot(x= X['number_courses'], y=error)
plt.show()
```

```python
sns.scatterplot(x= X['time_study'], y=error)
plt.show()
```

```python
sns.displot(error,  kind="kde")
plt.show()
```

```python
X.columns
```

```python
X['time_study_squared'] = X['time_study']*X['time_study']
```

```python
X.head()
```

```python
lm = LinearRegression()
lm.fit(X, y)

y_pred = lm.predict(X)
error = y-y_pred

sns.displot(error,  kind="kde")
plt.show()
```

```python
sns.scatterplot(x= X['number_courses'], y=error)
plt.show()
```

```python
sns.scatterplot(x= X['time_study_squared'], y=error)
plt.show()
```
