```python
import numpy as np
import pandas as pd

smsdata = pd.read_table('https://raw.githubusercontent.com/viveksharma-niet/LogisticRegressionUpgrade/refs/heads/main/smsspamcollection.csv', sep='^([^,]+),',
    engine='python',header=None, names=['Class', 'sms'], encoding="utf-8" )
smsdata.head()
```

```python
smsdata = smsdata[(smsdata.Class =='ham') | ( smsdata.Class =='spam') ]
```

```python
from sklearn.model_selection import train_test_split

smsdata['label'] = smsdata.Class.map({'ham':0, 'spam':1})
smsdata.head()
```

```python
X = smsdata.sms
y = smsdata.label
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, test_size=0.3, random_state=100)
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vec = CountVectorizer(stop_words='english')
vec.fit(X_train)
```

```python
X_train_tranformed = vec.transform(X_train)
X_test_tranformed = vec.transform(X_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
mnb = MultinomialNB()
mnb.fit(X_train_tranformed, y_train)


print(mnb.predict(X_test_tranformed))
print(mnb.predict_proba(X_test_tranformed))
```

```python
y_test_pred = mnb.predict_proba(X_test_tranformed)
```

```python
y_pred_class = mnb.predict(X_test_tranformed)
```

```python
from sklearn import metrics
```

```python
metrics.accuracy_score(y_test, y_pred_class)
```

```python
metrics.confusion_matrix(y_test, y_pred_class)
```

```python
metrics.precision_score(y_test, y_pred_class)
```
